# Comparison of MHSampler.jl with Turing.jl MH Sampler

This notebook aims to compare MHSampler.jl package with Turing.jl MH sampler

### Use packages

```julia
using Distributions
using Plots
using MHSampler
using Turing
using Random
Random.seed!(12345);
```

### Initialize data

```julia
l_w = 5
M = 3
W_mean = collect(range(0.0,2.0, length=l_w))
input = rand(l_w)
fn(ps) = input.*ps
output = fn(W_mean)
W_mean
```

### Initialize Likelihood function for MHSampler.jl

```julia
model(prm) = Normal.((fn(prm)), 5.0)
```

### Define prioirs and proposals

```julia
proposald = MvNormal(zeros(l_w), 5.0)
prior = MvNormal(zeros(l_w), 5.0)
proposalf() = rand(proposald)
```

### Sample using MHSampler.jl

```julia
chm = mh(prior, proposalf, model = model, output = output, itr =10_000, burn_in =0)
```

### Define Turing model

```julia

@model function gdemo(y, l_w)
  s ~ MvNormal(zeros(l_w), 5.0)
  op = fn(s)
  y ~ MvNormal(op, 5.0)
end
```

### Generate samples using Turing sampler

```julia
chn = sample(gdemo(output, l_w), MH(), 10_000)
```

```julia
#param1
plot(histogram(chn["s[1]"].data, title="Turing.jl"), histogram(Array(chm[1,2:end]), title="MHSampler.jl"), layout=2)
```

```julia
#param2
plot(histogram(chn["s[2]"].data, title="Turing.jl"), histogram(Array(chm[2,2:end]), title="MHSampler.jl"), layout=2)
```

### Mean and Standard Deviation

```julia
#param1
using LinearAlgebra
@show mean(chn["s[1]"].data) mean(chm[1,2:end])
@show std(chn["s[1]"].data) std(chm[1,2:end])
```
